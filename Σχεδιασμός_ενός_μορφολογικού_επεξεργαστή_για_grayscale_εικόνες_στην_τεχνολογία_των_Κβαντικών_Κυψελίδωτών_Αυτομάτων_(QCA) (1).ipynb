```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def load_image(path):
    img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
    return img

def show_image(img, title="Εικόνα"):
    plt.imshow(img, cmap='gray')
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
def morphological_ops(img, operation, kernel_size=3):
    kernel = np.ones((kernel_size, kernel_size), np.uint8)

    if operation == 'dilation':
        return cv2.dilate(img, kernel, iterations=1)
    elif operation == 'erosion':
        return cv2.erode(img, kernel, iterations=1)
    elif operation == 'opening':
        return cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
    elif operation == 'closing':
        return cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)
    else:
        raise ValueError("Unsupported operation")
```

```python
def load_image(path):
    img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
    if img is None:
        raise FileNotFoundError(f"Δεν βρέθηκε το αρχείο: {path}")
    return img

def show_image(img, title="Εικόνα"):
    plt.imshow(img, cmap='gray')
    plt.title(title)
    plt.axis('off')
    plt.show()

def morphological_ops(img, operation, kernel_size=3):
    kernel = np.ones((kernel_size, kernel_size), np.uint8)

    if operation == 'dilation':
        return cv2.dilate(img, kernel, iterations=1)
    elif operation == 'erosion':
        return cv2.erode(img, kernel, iterations=1)
    elif operation == 'opening':
        return cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
    elif operation == 'closing':
        return cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)
    else:
        raise ValueError("Unsupported operation")
```

```python
# Ανέβασε πρώτα τη σωστή εικόνα
img = load_image("Στιγμιότυπο οθόνης 2025-05-14 123130.png")
show_image(img, "Αρχική Εικόνα")

# Εφαρμογή μορφολογικών πράξεων
dilated = morphological_ops(img, 'dilation')
show_image(dilated, "Διαστολή")

eroded = morphological_ops(img, 'erosion')
show_image(eroded, "Διάβρωση")

opened = morphological_ops(img, 'opening')
show_image(opened, "Άνοιγμα")

closed = morphological_ops(img, 'closing')
show_image(closed, "Κλείσιμο")
```
