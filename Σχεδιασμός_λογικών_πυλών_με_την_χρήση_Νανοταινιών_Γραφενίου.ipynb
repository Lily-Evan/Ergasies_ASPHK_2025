```python
import numpy as np
import matplotlib.pyplot as plt

# Απλοποιημένη συνάρτηση που προσομοιώνει την αγωγιμότητα της νανοταινίας γραφενίου
def graphene_nanotape_conductivity(input1, input2, gate_voltage):
    """
    Επιστρέφει την αγωγιμότητα για μια νανοταινία γραφενίου
    Οι είσοδοι input1, input2 καθορίζουν τις τιμές των ηλεκτρικών πεδίων
    gate_voltage είναι το εφαρμοζόμενο εξωτερικό ηλεκτρικό πεδίο
    """
    # Εξαρτάται από τις εισόδους και την τάση του ηλεκτρικού πεδίου
    conductivity = (input1 * input2) * gate_voltage  # Απλή προσομοίωση
    return conductivity

# Συνάρτηση που υλοποιεί την λογική πύλη AND
def AND_gate(input1, input2, gate_voltage):
    conductivity = graphene_nanotape_conductivity(input1, input2, gate_voltage)
    return 1 if conductivity > 0.5 else 0  # Αν η αγωγιμότητα είναι υψηλή, έξοδος 1

# Συνάρτηση που υλοποιεί την λογική πύλη OR
def OR_gate(input1, input2, gate_voltage):
    conductivity = graphene_nanotape_conductivity(input1, input2, gate_voltage)
    return 1 if conductivity > 0 else 0  # Αν η αγωγιμότητα είναι μη μηδενική, έξοδος 1

# Συνάρτηση που υλοποιεί την λογική πύλη NOT
def NOT_gate(input1, gate_voltage):
    conductivity = graphene_nanotape_conductivity(input1, 1, gate_voltage)
    return 0 if conductivity > 0.5 else 1  # Αν η αγωγιμότητα είναι υψηλή, έξοδος 0, αλλιώς 1

# Συνάρτηση που υλοποιεί την λογική πύλη NAND
def NAND_gate(input1, input2, gate_voltage):
    conductivity = graphene_nanotape_conductivity(input1, input2, gate_voltage)
    return 0 if conductivity > 0.5 else 1  # Αν η αγωγιμότητα είναι υψηλή, έξοδος 0, αλλιώς 1

# Συνάρτηση που υλοποιεί την λογική πύλη NOR
def NOR_gate(input1, input2, gate_voltage):
    conductivity = graphene_nanotape_conductivity(input1, input2, gate_voltage)
    return 1 if conductivity == 0 else 0  # Αν η αγωγιμότητα είναι μηδενική, έξοδος 1

# Δοκιμή των πυλών με παραδείγματα εισόδων
inputs = [(0, 0), (0, 1), (1, 0), (1, 1)]
gate_voltage = 1.0  # Εφαρμοζόμενο ηλεκτρικό πεδίο

# Εμφάνιση αποτελεσμάτων για τις βασικές πυλές
print("AND Gate:")
for input1, input2 in inputs:
    print(f"Entrada: {input1}, {input2} -> Output: {AND_gate(input1, input2, gate_voltage)}")

print("\nOR Gate:")
for input1, input2 in inputs:
    print(f"Entrada: {input1}, {input2} -> Output: {OR_gate(input1, input2, gate_voltage)}")

print("\nNOT Gate:")
for input1, _ in inputs:
    print(f"Entrada: {input1} -> Output: {NOT_gate(input1, gate_voltage)}")

print("\nNAND Gate:")
for input1, input2 in inputs:
    print(f"Entrada: {input1}, {input2} -> Output: {NAND_gate(input1, input2, gate_voltage)}")

print("\nNOR Gate:")
for input1, input2 in inputs:
    print(f"Entrada: {input1}, {input2} -> Output: {NOR_gate(input1, input2, gate_voltage)}")

# Οπτικοποίηση της λειτουργίας των πυλών με δύο εισόδους
def plot_two_input_gate_operation(gate_function, title):
    results = np.array([[gate_function(input1, input2, gate_voltage) for input1, input2 in inputs]])
    plt.imshow(results, cmap="binary", aspect="auto")
    plt.title(title)
    plt.xticks([0, 1], ["0", "1"])
    plt.yticks([0], ["Outputs"])
    plt.xlabel("Input 2")
    plt.ylabel("Outputs")
    plt.show()

# Οπτικοποίηση της λειτουργίας πυλών με μία είσοδο (όπως NOT)
def plot_one_input_gate_operation(gate_function, title):
    # We only need the first element of each tuple in inputs
    single_inputs = [input1 for input1, _ in inputs]
    # Get unique inputs for the x-axis labels
    unique_inputs = sorted(list(set(single_inputs)))
    results = np.array([[gate_function(input1, gate_voltage) for input1 in unique_inputs]])

    plt.imshow(results, cmap="binary", aspect="auto")
    plt.title(title)
    plt.xticks(range(len(unique_inputs)), [str(i) for i in unique_inputs])
    plt.yticks([0], ["Outputs"])
    plt.xlabel("Input")
    plt.ylabel("Outputs")
    plt.show()


# Εμφάνιση γραφημάτων για τις πυλές
plot_two_input_gate_operation(AND_gate, "AND Gate")
plot_two_input_gate_operation(OR_gate, "OR Gate")
# Χρησιμοποιούμε τη νέα συνάρτηση για την NOT πύλη
plot_one_input_gate_operation(NOT_gate, "NOT Gate")
plot_two_input_gate_operation(NAND_gate, "NAND Gate")
plot_two_input_gate_operation(NOR_gate, "NOR Gate")
```
